```python
import pandas as pd
import altair as alt
```

### Generate Combined Datasets

```python
#initiate empty datasets
awhrs = pd.DataFrame()
asats = pd.DataFrame()
asats_2 = pd.DataFrame()
kendall = pd.DataFrame()
```

```python
# adding Matrix Factorization
awhrs_mf = pd.read_csv('Matrix Factorization/results/MF_awhrs.csv').drop(columns='Unnamed: 0')
asats_mf = pd.read_csv('Matrix Factorization/results/MF_asats.csv').drop(columns='Unnamed: 0')
asats_2_mf = pd.read_csv('Matrix Factorization/results/MF_asats2.csv').drop(columns='Unnamed: 0')
kendall_mf = pd.read_csv('Matrix Factorization/results/MF_Kendall.csv').drop(columns='Unnamed: 0')

awhrs_mf['Algorithm'] = 'Matrix Factorization'
asats_mf['Algorithm'] = 'Matrix Factorization'
asats_2_mf['Algorithm'] = 'Matrix Factorization'
kendall_mf['Algorithm'] = 'Matrix Factorization'

awhrs = pd.concat([awhrs, awhrs_mf], ignore_index=True)
asats = pd.concat([asats, asats_mf], ignore_index=True)
asats_2 = pd.concat([asats_2, asats_2_mf], ignore_index=True)
kendall = pd.concat([kendall, kendall_mf], ignore_index=True)
```

```python
# adding Deep FM
awhrs_dfm = pd.read_csv('DeepFM/results/DeepFM_awhrs.csv').drop(columns='Unnamed: 0')
asats_dfm = pd.read_csv('DeepFM/results/DeepFM_asats.csv').drop(columns='Unnamed: 0')
asats_2_dfm = pd.read_csv('DeepFM/results/DeepFM_asats2.csv').drop(columns='Unnamed: 0')
kendall_dfm = pd.read_csv('DeepFM/results/DeepFM_Kendall.csv').drop(columns='Unnamed: 0')

awhrs_dfm['Algorithm'] = 'DeepFM'
asats_dfm['Algorithm'] = 'DeepFM'
asats_2_dfm['Algorithm'] = 'DeepFM'
kendall_dfm['Algorithm'] = 'DeepFM'

awhrs = pd.concat([awhrs, awhrs_dfm], ignore_index=True)
asats = pd.concat([asats, asats_dfm], ignore_index=True)
asats_2 = pd.concat([asats_2, asats_2_dfm], ignore_index=True)
kendall = pd.concat([kendall, kendall_dfm], ignore_index=True)
```

```python
# adding AutoRec
awhrs_ar = pd.read_csv('AutoRec/results/AutoRec_awhrs.csv').drop(columns='Unnamed: 0')
asats_ar = pd.read_csv('AutoRec/results/AutoRec_asats.csv').drop(columns='Unnamed: 0')
asats_2_ar = pd.read_csv('AutoRec/results/AutoRec_asats2.csv').drop(columns='Unnamed: 0')
kendall_ar = pd.read_csv('AutoRec/results/AutoRec_Kendall.csv').drop(columns='Unnamed: 0')

awhrs_ar['Algorithm'] = 'AutoRec'
asats_ar['Algorithm'] = 'AutoRec'
asats_2_ar['Algorithm'] = 'AutoRec'
kendall_ar['Algorithm'] = 'AutoRec'

awhrs = pd.concat([awhrs, awhrs_ar], ignore_index=True)
asats = pd.concat([asats, asats_ar], ignore_index=True)
asats_2 = pd.concat([asats_2, asats_2_ar], ignore_index=True)
kendall = pd.concat([kendall, kendall_ar], ignore_index=True)
```

```python
# adding Caser
awhrs_c = pd.read_csv('Caser/results/Caser_awhrs.csv').drop(columns='Unnamed: 0')
asats_c = pd.read_csv('Caser/results/Caser_asats.csv').drop(columns='Unnamed: 0')
asats_2_c = pd.read_csv('Caser/results/Caser_asats2.csv').drop(columns='Unnamed: 0')
kendall_c = pd.read_csv('Caser/results/Caser_Kendall.csv').drop(columns='Unnamed: 0')

awhrs_c['Algorithm'] = 'Caser'
asats_c['Algorithm'] = 'Caser'
asats_2_c['Algorithm'] = 'Caser'
kendall_c['Algorithm'] = 'Caser'

awhrs = pd.concat([awhrs, awhrs_c], ignore_index=True)
asats = pd.concat([asats, asats_c], ignore_index=True)
asats_2 = pd.concat([asats_2, asats_2_c], ignore_index=True)
kendall = pd.concat([kendall, kendall_c], ignore_index=True)
```

```python
# adding random recommendations
awhrs_r = pd.read_csv('random/results/random_awhrs.csv').drop(columns='Unnamed: 0')
asats_r = pd.read_csv('random/results/random_asats.csv').drop(columns='Unnamed: 0')
asats_2_r = pd.read_csv('random/results/random_asats2.csv').drop(columns='Unnamed: 0')
kendall_r = pd.read_csv('random/results/random_Kendall.csv').drop(columns='Unnamed: 0')

awhrs_r['Algorithm'] = 'random'
asats_r['Algorithm'] = 'random'
asats_2_r['Algorithm'] = 'random'
kendall_r['Algorithm'] = 'random'

awhrs = pd.concat([awhrs, awhrs_r], ignore_index=True)
asats = pd.concat([asats, asats_r], ignore_index=True)
asats_2 = pd.concat([asats_2, asats_2_r], ignore_index=True)
kendall = pd.concat([kendall, kendall_r], ignore_index=True)
```

### Average Weighted Hitrate

```python
awhrs_line = alt.Chart(awhrs).mark_line().encode(
    x=alt.X('k:Q', title='Number of recommendations'),  # Quantitative scale for 'k'
    y='Average Weigthed Hit Rate:Q',  # Quantitative scale for 'Average Weighted Hit Rate'
    color='Algorithm:N'  # Nominal scale for different algorithms
)

awhrs_point = alt.Chart(awhrs).mark_point(size=100, filled=True, opacity=1).encode(
    x='k:Q',
    y='Average Weigthed Hit Rate:Q',
    color='Algorithm:N'
)


awhrs_chart = (awhrs_line + awhrs_point).properties(
    title='Average Weighted Hit Rate by Algorithm for Different Numbers of Recommendations',
    width=600,  # Width of the chart
    height=400  # Height of the chart
)
awhrs_chart
```

### Share of Users with at Least One Good Recommendation

```python
asats_line = alt.Chart(asats).mark_line().encode(
    x=alt.X('k:Q', title='Number of recommendations'),  # Quantitative scale for 'k'
    y=alt.Y('Average User Satisfaction:Q', title='Share of Users'),  # Quantitative scale for 'Average Weighted Hit Rate'
    color='Algorithm:N'  # Nominal scale for different algorithms
)

asats_point = alt.Chart(asats).mark_point(size=100, filled=True, opacity=1).encode(
    x='k:Q',
    y='Average User Satisfaction:Q',
    color='Algorithm:N'
)


asats_chart = (asats_line + asats_point).properties(
    title='Share of Users with at Least One 4 Star Recommendation',
    width=300,  # Width of the chart
    height=400  # Height of the chart
)
asats_chart
```

```python
asats_2_line = alt.Chart(asats_2).mark_line().encode(
    x=alt.X('k:Q', title='Number of recommendations'),  # Quantitative scale for 'k'
    y=alt.Y('Average User Satisfaction:Q', title='Share of Users'),  # Quantitative scale for 'Average Weighted Hit Rate'
    color='Algorithm:N'  # Nominal scale for different algorithms
)

asats_2_point = alt.Chart(asats_2).mark_point(size=100, filled=True, opacity=1).encode(
    x='k:Q',
    y='Average User Satisfaction:Q',
    color='Algorithm:N'
)


asats_2_chart = (asats_2_line + asats_2_point).properties(
    title='Share of Users with at Least One 5 Star Recommendation',
    width=300,  # Width of the chart
    height=400  # Height of the chart
)
asats_chart | asats_2_chart
```

### Kendall Distance with Penalty

```python
kendall_chart = alt.Chart(kendall).mark_bar().encode(
    x=alt.X('p:N', sort=['0.05', '0.2'], axis=alt.Axis(title='', labels=False)),  # Categorical scale for 'Algorithm', sorted by 'Kendall Distance'
    y='Kendall Distance:Q',  # Quantitative scale for 'Kendall Distance'
    color='p:N',  # Different colors for each 'p' value
    column=alt.Column('Algorithm:N', header=alt.Header(title='', orient='bottom'), sort='descending'),  # Separate bars for each 'p' value within each algorithm
    tooltip=['Algorithm', 'Kendall Distance', 'p']  # Show details on hover
).properties(
    title='Kendall Distance by Algorithm and p-Value',
    width=100,  # Width of each bar chart
    height=300  # Height of the chart
).configure_title(
    anchor='middle',
    align='center'  # Align the title to the center
)

kendall_chart.configure_facet(spacing=15).configure_view(stroke=None)
```
